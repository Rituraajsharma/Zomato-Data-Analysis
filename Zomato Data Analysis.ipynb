# Import all important libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Create the dataframe

```python
df = pd.read_csv("/content/drive/MyDrive/Zomato EDA/Zomato data .csv")
df
```

```python
df
```

#Convert the data type of column - rate

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
df.info()
```

# What type of resturant do the majority of customers order from

```python
 df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("type of resturant")
```

# Conclusion - the majority of the restaurant falls in dining category

# How many votes has each type of restaurant received from customers

```python
df.head()
```

```python
data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': data})
plt.plot(result,c="green",marker='o')
plt.xlabel("Type of restaurant",c='red',size=20)
plt.ylabel("votes",c='red',size=20)
```

#Conclusion :- Dinning restaurant recieves maximum votes

# What are the rating that majority of restaurant recieves

```python
df.head()
```

```python
plt.hist(df['rate'],bins=5)
plt.title("ratings of distribution")
plt.show()
```

# Conclusion - the majority restaurant recieved ratings 3.5 to 4

# Which mode recieved the maximum rating (online/offline)

```python
df.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=df)
```

Conclusion:- offline order recieved lower rating in comparison to online order

```python
df.head()
```

```python
pivot_table=df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("online_order")
plt.ylabel("Listed_in(type)")
plt.show()
```

Conclusion:- Dining restaurant primarily accept offline orders, whereas cafes primarily recieve online orders.

This suggest that client prefers orders in person at restaurant but prefer onlie orders at cafes.
